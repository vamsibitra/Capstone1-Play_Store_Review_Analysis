##Data Cleaning

````python
def convert_float(a):
  a = str(a)
  for i in a:
    if i in ',+Mk$':
      if i == 'k':
        a = str(a).replace(i,'')
        a = eval(a)/1024.0
      a = str(a).replace(i,'')
  if a == 'Free' or a =='Varies with device'or a =='Everyone':
    a=0
  else:
    a=eval(str(a))
  return a
````

````python
def convert_Million(a):
  if 'M' in a:
    a = str(a).replace('M', '')
    a = eval(str(a)) * 1000000
  return a
````

````python
#convert installs,price,reviews to float value so that it better for calculation
new_ps_data['Installs'] = new_ps_data['Installs'].apply(convert_float)
new_ps_data['Price'] = new_ps_data['Price'].apply(convert_float)
new_ps_data['Size'] = new_ps_data['Size'].apply(convert_float)
````

````python
new_ps_data['Reviews'] = new_ps_data['Reviews'].apply(convert_Million)
````

````python
# Chart - 2 top 10 installed apps in any category
def find_top_10_in_category(category, data):
    category = category.upper()
    top_10_apps = data[data['Category'] == category].nlargest(10, 'Installs')

    plt.figure(figsize=(15, 12))
    plt.title(f'Top 10 Installed Apps in {category} Category', size=20)
    ax = sns.barplot(x='Installs', y='App', data=top_10_apps)
    ax.set(xlabel='Number of Installs', ylabel='App Name')
    plt.xticks(rotation=45, ha='right')
    plt.show()
find_top_10_in_category('Communication', new_ps_data1)
````

##### 1. Why did you pick the specific chart?

It visualizes the top 10 installed apps in any category

##### 2. What is/are the insight(s) found from the chart?

The graph above reveals that within the Communication category, the apps with the highest number of installations are Messenger - Text and Video Chat for Free, WhatsApp Messenger, and Gmail.
